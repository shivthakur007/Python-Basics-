### Loops


```python
#printing a string 5 times using for loop
for i in range(5):
    print("Hello World")
```

```python
#printing a string 5 times using while loop
i = 1
while i <= 5:
    print("Hello World")
    i += 1
    
```

```python
#iterating over list
my_list = ["Apple", "Banana", "Orange"]
for fruits in my_list:
    print(fruits)
```

```python
#iterating over a range
n = int(input())
for i in range(0, n+1, 2): #(start, end, step)
    print(i)

#the end is always exclusive so if you want 10 at the end write 11
```

```python
#iterating over a string
name = "ShivThakur"
for character in name:
    print(character)
```

```python
#iterating over a string and you want both index and character 
name = "ShivThakur"
for index, character in enumerate(name):
    print(index, character)
```

```python
#iterating over dictionary
#Dictionary has key-value pair
student = {
    "Name" : "Shiv",
    "Age" : 27,
    "Course" : "Data Science"
}

for key in student: #for printing key
    print(key)
    
for value in student.values(): #for printing values
    print(value)

for key, value in student.items():
    print(key ,":", value)
```

```python
#print all even numbers between numbers
n = int(input("Enter the Start number: " ))
i = int(input("Enter the Lastnumber: " ))
while n <= i:
    if n % 2 == 0:
        print(n)
    n += 1
```
